**Question 2:** In how many different states do the customers live in?

```python
with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Creamos un conjunto vacío para guardar los estados sin que se repitan
    states = set()

    for row in csvreader:
        # El estado está en la quinta columna (índice 4 en Python)
        state = row[4]
        states.add(state)

# Contamos cuántos elementos quedaron en el conjunto
print(f"Different states: {len(states)}")# evaluate the contents of the file
```

**Question 3** What is the state with most customers?

```python
with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Diccionario para contar clientes por estado
    state_counts = {}

    for row in csvreader:
        state = row[4]
        # Si el estado ya está en el diccionario, sumamos 1. Si no, lo inicializamos en 0 y sumamos 1.
        state_counts[state] = state_counts.get(state, 0) + 1

# Buscamos el estado con el conteo máximo
max_state = max(state_counts, key=state_counts.get)
max_customers = state_counts[max_state]

print(f"State with most customers: {max_state} ({max_customers} customers)")# evaluate the contents of the file
```

**Question 4** What is the state with the least customers?

```python
with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Diccionario para contar clientes por estado
    state_counts = {}

    for row in csvreader:
        state = row[4]
        state_counts[state] = state_counts.get(state, 0) + 1

# Buscamos el estado con el conteo mínimo
min_state = min(state_counts, key=state_counts.get)
min_customers = state_counts[min_state]

print(f"State with least customers: {min_state} ({min_customers} customers)")# evaluate the contents of the file
```

**Question 5:** What is the most common last name?

```python
with open(datafile, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Diccionario para contar los apellidos
    lastname_counts = {}

    for row in csvreader:
        last_name = row[2].strip() # .strip() por si hay espacios extra
        lastname_counts[last_name] = lastname_counts.get(last_name, 0) + 1

# Buscamos el apellido más común
most_common_lastname = max(lastname_counts, key=lastname_counts.get)
count = lastname_counts[most_common_lastname]

print(f"Most common last name: {most_common_lastname} ({count} times)")# evaluate the contents of the file
```

```python
import csv

orders_file = "./sample_data/orders.csv"

with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Conjunto para guardar los IDs de órdenes únicas
    unique_orders = set()

    for row in csvreader:
        # El OrderID está en la segunda columna (índice 1)
        order_id = row[1]
        unique_orders.add(order_id)

print(f"Unique orders: {len(unique_orders)}")
```

```python
with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Contamos la cantidad total de filas (ítems)
    total_items = 0
    for row in csvreader:
        total_items += 1

# Calculamos el promedio usando las 16672 órdenes únicas
average_items = total_items / 16672

print(f"Total items: {total_items}")
print(f"Average items per order: {average_items:.2f}")
```

```python
with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Diccionario para contar cuántos ítems tiene cada OrderID
    order_counts = {}

    for row in csvreader:
        order_id = row[1]
        order_counts[order_id] = order_counts.get(order_id, 0) + 1

# Buscamos la orden con el conteo máximo
max_order = max(order_counts, key=order_counts.get)
max_items = order_counts[max_order]

print(f"Order with most items: {max_order} ({max_items} items)")
```

```python
with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    max_price = 0.0
    expensive_product = ""

    for row in csvreader:
        # El precio está en la sexta columna (índice 5)
        price = float(row[5])
        product_name = row[4]

        # Si encontramos un precio mayor al que teníamos, lo actualizamos
        if price > max_price:
            max_price = price
            expensive_product = product_name

print(f"Most expensive item: {expensive_product} (Price: ${max_price})")
```

```python
with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Diccionario para contar la cantidad de veces que se vende cada producto
    product_counts = {}

    for row in csvreader:
        product_name = row[4]
        product_counts[product_name] = product_counts.get(product_name, 0) + 1

# Buscamos el producto más vendido
most_sold_product = max(product_counts, key=product_counts.get)
times_sold = product_counts[most_sold_product]

print(f"Most sold item: {most_sold_product} ({times_sold} times)")
```

```python
import csv

with open(orders_file, 'r') as fl:
    csvreader = csv.reader(fl, delimiter=',')
    header = next(csvreader) # Saltamos el encabezado

    # Diccionario para guardar el total de cada orden única y saber su mes
    # Estructura: {order_id: (mes, orden_total)}
    unique_orders = {}

    for row in csvreader:
        order_id = row[1]
        date_str = row[2]
        order_total = float(row[3])

        # Extraemos el mes de la fecha (formato YYYY-MM-DD...)
        # Tomamos los caracteres del índice 5 al 7 (el MM)
        month = date_str[5:7]

        if order_id not in unique_orders:
            unique_orders[order_id] = (month, order_total)

# Ahora sumamos los totales agrupando por mes
month_sales = {}
for month, total in unique_orders.values():
    month_sales[month] = month_sales.get(month, 0.0) + total

# Buscamos el mes con mayores ventas
best_month = max(month_sales, key=month_sales.get)
max_sales = month_sales[best_month]

# Diccionario para mostrar el nombre del mes
month_names = {
    "01": "January", "02": "February", "03": "March", "04": "April",
    "05": "May", "06": "June", "07": "July", "08": "August",
    "09": "September", "10": "October", "11": "November", "12": "December"
}

print(f"Historically, the best month for sales is: {month_names[best_month]} ({best_month}) with ${max_sales:,.2f}")
```
